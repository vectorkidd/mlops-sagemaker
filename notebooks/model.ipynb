```python
import os
import pandas as pd
from sklearn.model_selection import train_test_split

data_path = "/home/sagemaker-user/mlops-sagemaker/data/raw_data.csv"

df = pd.read_csv(data_path)

df.head
```

```python
df.shape
```

```python
os.makedirs("data", exist_ok=True)

X = df[['cgpa','iq','profile_score']]
y = df['placed']

X_train, X_test, y_train, y_test = train_test_split(X,y , random_state=42, test_size=0.2) 

train_df = X_train.copy() 
train_df['placed'] = y_train 

test_df = X_test.copy() 
test_df['placed'] = y_test

train_path = "/home/sagemaker-user/mlops-sagemaker/data/train_data.csv"
test_path = "/home/sagemaker-user/mlops-sagemaker/data/test_data.csv"

train_df.to_csv(train_path, index=False)
test_df.to_csv(test_path, index=False)

train_df.shape
test_df.shape
```

```python
import mlflow
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
import pandas as pd

train_path = "/home/sagemaker-user/mlops-sagemaker/data/train_data.csv"
test_path = "/home/sagemaker-user/mlops-sagemaker/data/test_data.csv"

train_df = pd.read_csv(train_path)
test_df = pd.read_csv(test_path)

X_train = train_df.drop("placed", axis=1)
y_train = train_df["placed"]

X_test = test_df.drop("placed", axis=1)
y_test = test_df["placed"]

with mlflow.start_run():

    mlflow.autolog()

    model = LogisticRegression()


    model.fit(X_train, y_train)

    y_pred = model.predict(X_test)

    acc = accuracy_score(y_test, y_pred)
    
```
